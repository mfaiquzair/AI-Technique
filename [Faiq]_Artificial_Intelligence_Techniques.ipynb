```python
# Import libraries from skfuzzy
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# Declaring input and output
carat = ctrl.Antecedent(np.arange(0, 5.1, 0.05), 'carat')
depth = ctrl.Antecedent(np.arange(0, 101, 1), 'depth')
table = ctrl.Antecedent(np.arange(0, 101, 0.5), 'table')
price = ctrl.Consequent(np.arange(0, 21, 1), 'price')
```

```python
# Carat membership function
carat.automf(3)
```

```python
# Depth membership function
depth['shallow'] = fuzz.trapmf(depth.universe, [0, 0, 50, 60])
depth['ideal'] = fuzz.trimf(depth.universe, [50, 60, 70])
depth['deep'] = fuzz.trapmf(depth.universe, [60, 70, 100, 100])
```

```python
# Table membership function
table['narrow'] = fuzz.trapmf(table.universe, [0, 0, 50, 57.5])
table['ideal'] = fuzz.trimf(table.universe, [50, 57.5, 65])
table['wide'] = fuzz.trapmf(table.universe, [57.5, 65, 101, 101])
```

```python
# Price membership function
price['less expensive'] = fuzz.trimf(price.universe, [0, 0, 10])
price['expensive'] = fuzz.trimf(price.universe, [0, 10, 20])
price['more expensive'] = fuzz.trimf(price.universe, [10, 20, 20])
```

```python
# Membership function in graphic
carat.view()
depth.view()
table.view()
price.view()
```

Fuzzy Rules

```python
# Fuzzy rules
rule1 = ctrl.Rule(carat['good'] & depth['ideal'] & table['ideal'], price['more expensive'])
rule2 = ctrl.Rule(carat['good'] & (depth['ideal'] | table['ideal']), price['more expensive'])
rule3 = ctrl.Rule(carat['good'], price['expensive'])
rule4 = ctrl.Rule(carat['average'] & depth['ideal'] & table['ideal'], price['more expensive'])
rule5 = ctrl.Rule(carat['average'] & (depth['ideal'] | table['ideal']), price['expensive'])
rule6 = ctrl.Rule(carat['average'], price['less expensive'])
rule7 = ctrl.Rule(carat['poor'], price['less expensive'])
```

Fuzzy logic

```python
diamond_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
```

```python
diamond = ctrl.ControlSystemSimulation(diamond_ctrl)
```

```python
# Add new column to fuzzy logic output
data['newPrice'] = 0
data['accuracy'] = 0
```

```python
# Insert data from dataset, calculate the fuzzy logic, and add outputs to new column
for x in range(len(data)):
  # Insert data
  diamond.input['carat'] = data.iloc[x, 0]
  diamond.input['depth'] = data.iloc[x, 1]
  diamond.input['table'] = data.iloc[x, 2]

  # Calculate the fuzzy logic
  diamond.compute()

  # Add outpus to new column
  newPrice = round(diamond.output['price'] * 100)
  if newPrice <= data.iloc[x, 3]:
    accuracy = newPrice/data.iloc[x,3] * 100
  else:
    sub = newPrice - data.iloc[x, 3]
    accuracy = (data.iloc[x, 3] - sub)/data.iloc[x,3] * 100
  data['newPrice'][x] = newPrice
  data['accuracy'][x] = accuracy
```
